```python
from xml.dom.minidom import parse
import xml.dom.minidom

# Open XML document using minidom parser
DOMTree = xml.dom.minidom.parse("plos.correction.00a3b22e-36a9-4d51-89e5-1e6561e7a1e9.xml")
collection = DOMTree.documentElement
journaltitle = collection.getElementsByTagName('article')
#print(journaltitle.nodeName)
print(collection.nodeName)
print(collection.firstChild.nodeName)
for i in journaltitle:
    print(i)
#print(journaltitle.nodeName)
```

```python
import xml.etree.ElementTree as ET
tree = ET.parse('plos.correction.00a3b22e-36a9-4d51-89e5-1e6561e7a1e9.xml')
root = tree.getroot()
print(root.tag)
# all items data
for elem in root:
    print(elem.tag)
    for subelem in elem:
        for subielem in subelem:
            for suelem in subielem:
                for selem in suelem:
                    print(selem.tag)
                    print(selem.text)
                print(suelem.tag)
                print(suelem.text)
            print(subielem.tag)
            print(subielem.text)
        print(subelem.text)
        
```
